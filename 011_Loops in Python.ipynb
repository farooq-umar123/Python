## __Loop's in Python:__
- In python we have two types of loop's:
1. While loop.
2. For loop.

### __`While Loop:`__
- Python __`while`__ loop is used to execute a block of statements repeatedly until a given condition is satisfied.
- When the condition becomes false, the line immediately after the loop in the program is executed.
- Increment has to be given after statement in while loop orelse the loop continues and never ends.
- While loop should be closed by __`colon (:)`__.



___`Example:`___

```python
# simple program using while loop
num = 0
while num<=10:
    print(num)
    num += 1
```

```python
year = int(input("Enter Year:"))
while year<=2050:
    if year%100 == 0:
        print(year, "is", year//100, "Century")
    else:
        print(year, "is", year//100+1, "Century")
    year = year + 50
```

```python
num = 6
while num<=15:
    if num%2 == 0:
        print(num, "Is Even Number")
    else:
        print(num, "Is Odd Number")
    num = num + 1
```

```python
year = 2015
while year<=2028:
    if year%4 == 0:
        print(year, "Is a Leap Year")
    else:
        print(year, "Is not a Leap Year")
    year = year + 1
```

### __`For Loop:`__
- The For Loops in Python are a special type of loop statement that is used for sequential traversal.
- Python For loop is used for iterating over an iterable like a String, Tuple, List, Set, or Dictionary.



___`Example:`___

```python
print("List Iteration")
list = ["Apple", "Mango", "Cherry"]
for i in list:
    print(i)
```

```python
print("Tuple Iteration")
tup = ("Apple", 10, 1+2j, True)
for i in tup:
    print(i)
```

```python
print("String Iteration")
str = "Hello"
for i in str:
    print(i)
```

```python
print("Dictionary Iteration")
dict = {"Name:":"Umar", "Age:":25, "Location:":"Hyderabad"}
for x,y in dict.items():
    print(f"{x} {y}")
```

```python
print("Set Iteration")
set = {"Cherry", 10, 1+6j, 10, "BMW", "BMW"}
for i in set:
    print(i)
```

```python
vovcon = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
char = input("Enter Value:")
for a in char:
    if a in vovcon:
        print("Vowel", a)
    else:
        print("Consonent", a)
```

```python
n = 10
for i in range(0, n, 2):
    print(i)
```

```python
a = input("Enter Character:")
for i in a:
    if i>='A' and i<='Z':
        print("Given Character is UpperCase")
    else:
        print("Give Character is LowerCase")
```
